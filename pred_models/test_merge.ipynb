```python
import pandas as pd
import numpy as np
```

```python
#データの読み込み
df1 = pd.read_csv('../../data/features/feature_historical_1212-2212.csv', encoding='utf-8')
df2 = pd.read_csv('../../data/features/pl_all_500_1601-2212.csv', encoding='utf-8')
```

```python
# 'date' 列を 'Date' に変更して新しいデータフレームを作成
df1 = df1.rename(columns={'date': 'Date'})
# 'Date' 列を文字列に変換
df1['Date'] = df1['Date'].astype(str)
df1
```

```python
#シリアル値→日付
df2['Date'] = pd.to_datetime(df2['Date'], unit='D', origin='1899-12-30')
df2['Date'] = df2['Date'].astype(str)

# 日付のフォーマットを統一する
df2['Date'] = df2['Date'].str.replace('-', '')
#df2.drop('Invest', axis=1, inplace=True)
df2
```

```python
drop_col = ['label', 'Invest']
result_df = pd.merge(df1, df2, on='Date', how='outer')
result_df = result_df.drop(drop_col, axis=1)
```

```python
#儲けるか儲けられないかのlabelを追加し，不要な行の削除
result_df['label'] = (result_df['P/L'] > 0).astype(int)
result_df = result_df.dropna(subset=['P/L'])
result_df = result_df.drop('P/L', axis=1)
result_df.to_csv('../../data/features/feature_profit_1212-2212.csv', index=False, encoding='utf-8')
result_df
```
